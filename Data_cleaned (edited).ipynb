```python
df["Company Name"] = df["company_txt"]
```

```python
df=df.drop(columns={'company_txt','Age','Headquarters'})
```

```python
df = df.rename(columns={"job_title_sim":"Title Cat", "seniority_by_title":"Seniority","Job Location":"State"})
```

```python
df["Rating"]=df["Rating"].astype(str).str.replace("-1", "1").astype(float)
df["Founded"]=df["Founded"].astype(str).str.replace("-1", "Unknown")
df["Industry"]=df["Industry"].astype(str).str.replace("-1", "Unknown")
df["Sector"]=df["Sector"].astype(str).str.replace("-1", "Unknown")
df["Competitors"]=df["Competitors"].astype(str).str.replace("-1", "None")
df["Degree"]=df["Degree"].str.replace("na", "B")
df["Hourly"]=df["Hourly"].astype(str).str.replace("0", "Fulltime").str.replace("1","Parttime")
df[["Rating", "Founded", "Industry", "Sector", "Competitors", "Degree", "Hourly", "Title Cat", "Seniority", "State"]]
```

```python
def rank_rating(rate):
    if rate >= 1 and rate < 2:
        return "bad"
    elif rate >= 2 and rate < 3:
        return "normal"
    elif rate >= 3 and rate < 4:
        return "good"
    elif rate >= 4:
        return "great"

df["Rank"] = df["Rating"].apply(rank_rating)
df[["Rating", "Rank"]]
```
